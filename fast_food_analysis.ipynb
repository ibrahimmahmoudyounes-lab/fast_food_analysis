###Average Age per Gender Group

```python
Gender_Age=df.groupby('Gender')[['Age']].mean()
```

```python
Gender_Age
```

```python


plot_df = Gender_Age.reset_index()
ax = sns.barplot(data=plot_df, x='Gender', y='Age',hue='Gender' )

for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=3)

plt.ylim(0, plot_df['Age'].max() + 5)
plt.show()
```

The visualization depicts the average age for each gender group, revealing a high degree of uniformity. The data indicates that the mean age remains stable across all categories, with a narrow range between 37.15 and 39.03 years.

###Weekly Fast Food Consumption by Gender

```python
Gender_Fast_Food_Meals_Per_Week=df.groupby('Gender')[['Fast_Food_Meals_Per_Week']].mean()
```

```python
Gender_Fast_Food_Meals_Per_Week
```

```python
plt.pie(Gender_Fast_Food_Meals_Per_Week['Fast_Food_Meals_Per_Week'],
        labels = Gender_Fast_Food_Meals_Per_Week.index,
        autopct = '%1.1f%%',
        startangle = 90)

plt.axis('equal')
plt.show()
```

This pie chart visualizes the distribution of average weekly fast-food consumption among genders. The near-equal segments, ranging from 32.6% to 34.2%, demonstrate that fast-food habits are remarkably balanced across all groups, with no single gender showing a dominant trend.
